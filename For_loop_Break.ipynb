# Task 1: print number from1 to 5

```python
print("Task 1:")
for i in range(1,6):
    print(i)
```

# Task 2: print odd number from 1 to 10

```python
print("\nTask 2:")
for i in range(1,11, 2):
    print(i)
```

# Task 3: Nested loop pattern

```python
print("\nTask 3:")
for i in range(3):        # 0,1,2
    for j in range(2):  # 0,1
        print(i, j)
```

# Task 4: Break when number is 5

```python
print("\nTask 4:")
for i in range(1,8):
    if i == 5:
        break
    print(i)
```
