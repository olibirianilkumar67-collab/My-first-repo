Theatre Booking System

```python
Total_Seats = 350
available_seats = Total_Seats

total_bookings = 0
tickets_sold = 0
rejected_booking = 0

# Main booking loop
while True:
  print(f"\nAvailable Seats: {available_seats}")

  # Fix: Initial Indentation 1Error and typo 'in' -> 'int'
  tickets_input = input("Enter number of tickets (0 to exit):")

  # Ensure input is a valid integer before proceeding
  try:
    tickets = int(tickets_input)
  except ValueError:
    print("BOOKING REJECTED - Invalid input. Please enter a number.")
    rejected_booking += 1
    continue

  # Exit condition
  if tickets == 0:
    break

  # Validate ticket range (message corrected to match condition)
  if tickets < 1 or tickets > 15:
    print("BOOKING REJECTED - Invalid number of tickets (1-15 allowed)")
    rejected_booking += 1
    continue

  # Check seat availability
  if tickets > available_seats:
    print("BOOKING REJECTED - Not enough seats available")
    rejected_booking += 1
    continue

  # Age validation
  valid_age = True
  for i in range(tickets):
    age_input = input(f"Enter age of person {i+1}:")
    try:
      age = int(age_input)
    except ValueError:
      print("BOOKING REJECTED - Invalid age input. Please enter a number.")
      valid_age = False # Mark as invalid age
      break # Exit age loop
    if age < 12: # Assuming 12 is the minimum age based on typical restrictions
      valid_age = False
      break # Exit age loop if one person is underage

  # If age is invalid
  if not valid_age:
    print("BOOKING REJECTED - Age restriction violated (minimum age 12 required).")
    rejected_booking += 1
    continue

  # Booking successful
  print(f"Booking successful - {tickets} tickets")
  total_bookings += 1
  tickets_sold += tickets
  available_seats -= tickets

  # Stop if theatre is full
  if available_seats == 0:
    print("\nTheatre FULL")
    break

# Final Report (This should be outside the while loop)
print("\n===== Final Report=====")
print(f"Total Bookings: {total_bookings}")
print(f"Tickets Sold: {tickets_sold}")
print(f"Rejected Bookings:{rejected_booking}")
print(f"Remaining Seats: {available_seats}")
```
